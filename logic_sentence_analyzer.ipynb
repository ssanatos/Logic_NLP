```python
import re

sentence = "실패는 목적을 이루지 못한 것이다"
fix_sen = re.sub('한 것이다','하다',sentence)
print(fix_sen)
```

```python
from konlpy.tag import Kkma

kkma = Kkma()
sentence1 = "실패는 성공의 어머니이다.어떤 탈락은 실패이다.어떤 탈락은 성공의 어머니이다."

def divide_sentence(sentence):
    kkma = Kkma()
    sentence = re.sub('하다','한 것이다',sentence)
    words = kkma.pos(sentence)
    N_index = []
    cut_sent=0
    for idx, word in enumerate(words):
        if word[1].startswith('N'):  # N은 명사 J는 조사
            N_index.append(idx)
    for i in N_index:
        if len(words) == int(i+1):
            break
        if words[i+1][1].startswith('J'):
            if words[i+1][0] == '는' :
                search_str = words[i][0] + '는'
                cut_sent = sentence.find(search_str) + len(search_str) + 1
            elif words[i+1][0] == '은':
                search_str = words[i][0] + '은'
                cut_sent = sentence.find(search_str) + len(search_str) + 1
    return sentence[:cut_sent], sentence[cut_sent:]

def analyze_syllogism(syllogism):
    sentences = syllogism.split('.')
    divide_sent = []
    for sent in sentences:
        try:
            subject, predicate = divide_sentence(sent)
            if len(subject)> 1 and len(predicate) > 1 :
                sub_pos = kkma.pos(subject)
                print(sub_pos)
                sub_morphs = []
                for word in sub_pos:
                    if word[1][0] != 'E' and word[1][0] != 'I' and word[1][0] != 'S' and word[1][0] != 'J' and word[1] != 'VCP' and word[0] != '모든':
                        sub_morphs.append(word[0])
                divide_sent.append(sub_morphs)
                pre_pos = kkma.pos(predicate)
                print(pre_pos)
                pre_morphs = []
                for word in pre_pos:
                    if word[1][0] != 'E' and word[1][0] != 'I' and word[1][0] != 'S' and word[1][0] != 'J' and word[1] != 'VCP' and word[0] != '모든':
                        pre_morphs.append(word[0])
                divide_sent.append(pre_morphs)
        except:
            continue
    return divide_sent
```

```python
def nega_posi_condition(logic_form):
    neg = ['않','아니','못하']
    posi_nega=''
    for phrase in logic_form:
        if len(list(set(neg) - set(phrase))) == 3:
            posi_nega += 'P'
        else:
            posi_nega += 'N'
    return posi_nega

def include_condition(n,m,logic_form):
    if len(list(set(logic_form[n]) - set(logic_form[m]))) == 0:
        return True
    else:
        return False

def discriminant(logic_form):
    form_result = False
    posi_nega = nega_posi_condition(logic_form)
    if posi_nega=='PPPPPP':
        # 1번째
        if logic_form[0] == logic_form[3] and logic_form[1] == logic_form[5] and logic_form[2] == logic_form[4]:
            form_result = True
        # 1번째 전제 순서 뒤집은 거
        if logic_form[2] == logic_form[1] and logic_form[3] == logic_form[5] and logic_form[0] == logic_form[4]:
            form_result = True
        # 2번째
        if logic_form[0] == logic_form[2] and logic_form[1] == logic_form[5] and include_condition(3,4,logic_form):
            form_result = True
        # 2번째 전제 순서 뒤집은 거
        if logic_form[2] == logic_form[0] and logic_form[3] == logic_form[5] and include_condition(1,4,logic_form):
            form_result = True
        # 3번째
        if logic_form[0] == logic_form[5] and logic_form[1] == logic_form[2] and include_condition(3,4,logic_form):
            form_result = True
        # 3번째 전제 순서 뒤집은 거
        if logic_form[2] == logic_form[5] and logic_form[3] == logic_form[0] and include_condition(1,4,logic_form):
            form_result = True

    elif posi_nega=='PNPPPN':
        # 4번째
        if logic_form[0] == logic_form[3] and logic_form[1] == logic_form[5] and logic_form[2] == logic_form[4]:
            form_result = True
        # 5번째
        if logic_form[0] == logic_form[2] and logic_form[1] == logic_form[5] and include_condition(3,4,logic_form):
            form_result = True
        # 8번째 전제 순서 뒤집은 거
        if include_condition(2,5,logic_form) and include_condition(3,1,logic_form) and logic_form[0] == logic_form[4]:
            form_result = True
        # 9번째
        if include_condition(2,0,logic_form) and logic_form[1] == logic_form[5] and include_condition(3,4,logic_form):
            form_result = True

    elif posi_nega=='PPPNPN':
        # 4번째 전제 순서 뒤집은 거
        if logic_form[2] == logic_form[1] and logic_form[3] == logic_form[5] and logic_form[0] == logic_form[4]:
            form_result = True
        # 5번째 전제 순서 뒤집은 거
        if logic_form[2] == logic_form[0] and logic_form[3] == logic_form[5] and include_condition(1,4,logic_form):
            form_result = True
        # 8번째
        if include_condition(0,5,logic_form) and include_condition(1,3,logic_form) and logic_form[2] == logic_form[4]:
            form_result = True
        # 9번째 전제 순서 뒤집은 거
        if include_condition(0,2,logic_form) and logic_form[3] == logic_form[5] and include_condition(1,4,logic_form):
            form_result = True

    return form_result
```
